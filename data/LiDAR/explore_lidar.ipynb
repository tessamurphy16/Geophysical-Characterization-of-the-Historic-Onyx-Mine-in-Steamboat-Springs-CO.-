```python
import rasterio
from rasterio.plot import show
from glob import glob
import matplotlib.pyplot as plt
import numpy as np
```

```python
files = glob('LiDAR_2026-06-05T15_32_32.740Z/*.tif')
```

```python
fig, ax = plt.subplots(1,1)
for f in files:
    dem = rasterio.open(f)
    elev_data = dem.read(1)
    bounds = dem.bounds
    cb = ax.imshow(elev_data, cmap='Greys_r', extent = bounds, vmin=6650, vmax=7000)
#ax.set_xlim([2615000,2645000])
#ax.set_ylim([1377000,1407000])
ax.set_aspect('equal')
ax.set_xlabel('easting [ft]')
ax.set_ylabel('northing [ft]')
fig.colorbar(cb, ax = ax, label = 'elevation [ft]')
plt.show()
```
